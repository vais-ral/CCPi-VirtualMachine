# CCPi using Octave in computation

This notebook contains basic demo how to use CCPi with GNU Octave and/or MATLAB &trade;.
Before running this notebook: 
- copy `mex_compile` folder from https://github.com/vais-ral/CCPi-Regularisation-Toolkit/tree/master/src/Matlab/mex_compile into `notebooks/mex_compile`
- copy `data` folder from https://github.com/vais-ral/CCPi-Regularisation-Toolkit/tree/master/demos/data into `notebooks/data`


First, we create 3d plot to check Octave is working as expected:  

```octave
%plot --format png
tx = ty = linspace (-9, 9, 41)';
[xx, yy] = meshgrid (tx, ty);
r = sqrt (xx .^ 2 + yy .^ 2) + eps;
tz = sin (r) ./ r;
mesh (tx, ty, tz);
```

Now the following code expects that `mex_compile` and `data` folder were coppied from CCPi-Regularisation-Toolkit repository. 

TODO: if the code bellow fail on `support for Image IO was unavailable` then the Octave installed suffers with issue reported https://github.com/conda-forge/octave-feedstock/issues/18


```octave
% Volume (3D) denoising demo using CCPi-RGL
clear; close all
Path1 = sprintf(['mex_compile' filesep 'installed'], 1i);
Path2 = sprintf(['data' filesep], 1i);
addpath(Path1);
addpath(Path2);

N = 512; 
slices = 15;
vol3D = zeros(N,N,slices, 'single');
Ideal3D = zeros(N,N,slices, 'single');
Im = double(imread('data/lena_gray_512.tif'))/255;  % loading image
```

```octave
for i = 1:slices
vol3D(:,:,i) = Im + .05*randn(size(Im)); 
Ideal3D(:,:,i) = Im;
end
vol3D(vol3D < 0) = 0;
figure; imshow(vol3D(:,:,15), [0 1]); title('Noisy image');
```
